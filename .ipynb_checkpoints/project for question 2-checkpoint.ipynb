```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
spx = pd.read_csv("https://stooq.com/q/d/l/?s=es.c&i=d", parse_dates=["Date"])
vix = pd.read_csv("https://stooq.com/q/d/l/?s=vi.c&i=d", parse_dates=["Date"])

spx = spx.sort_values("Date")
vix = vix.sort_values("Date")
spx = spx.set_index("Date")
vix  = vix.set_index("Date")
end = spx.index.max()
start = end - pd.DateOffset(months=6)
spx = spx.loc[start:end]
vix  = vix.loc[start:end]
spx_close = spx["Close"]
R=spx_close.pct_change()
R_bar=R.rolling(window=30).mean()
sq_diff=(R-R_bar)**2
sum_sq=sq_diff.rolling(window=30).sum()
sigma_bar = 100 * np.sqrt(252/29 * sum_sq)
sigma_bar = sigma_bar.rename("hist_vol")
vix_close = vix["Close"].rename("VIX")
df = pd.concat([sigma_bar, vix_close], axis=1).dropna()
print(df.tail())
plt.figure(figsize=(10,5))
plt.plot(df.index, df["hist_vol"], label="30D realized vol (_i)")
plt.plot(df.index,df["VIX"],      label="VIX (v_i)")
plt.ylabel("Volatility (%)")
plt.title("VIX vs 30-day realized volatility of SPX (Stooq)")
plt.legend()
plt.grid(True)
plt.show()
```
